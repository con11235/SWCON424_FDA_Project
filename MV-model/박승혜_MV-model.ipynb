```python
symbols = []
while True:
    name = input("종목코드 : ")
    if name == '0':
        break
    symbols.append(name)
st = input("시작일 0000-00-00: ")
en = input("종료일 0000-00-00: ")
lookback = input("Look back : ")
rebalance = input("Rebalance : ")
```

```python
symbols = ['AAPL', 'MSFT', 'AMZN', 'FB']#, 'GOOGL', 'GOOG', 'TSLA', 'NVDA', 'CMCSA', 'ADBE']
st ='2020-09-01'
en = '2020-11-22'
lookback = '6m'
rebalance = '1m'
st = datetime(int(st[:4]),int(st[5:7]),int(st[8:10]))
en = datetime(int(en[:4]),int(en[5:7]),int(en[8:10]))
if lookback[-1] == 'd':
    lookback = timedelta(days=int(lookback[:-1]))
elif lookback[-1] == 'w':
    lookback = timedelta(weeks=int(lookback[:-1]))
elif lookback[-1] == 'm':
    lookback = relativedelta(months=int(lookback[:-1]))
elif lookback[-1] == 'y':
    lookback = relativedelta(years=int(lookback[:-1]))
if rebalance[-1] == 'd':
    rebalance = timedelta(days=int(rebalance[:-1]))
elif rebalance[-1] == 'w':
    rebalance = timedelta(weeks=int(rebalance[:-1]))
elif rebalance[-1] == 'm':
    rebalance = relativedelta(months=int(rebalance[:-1]))
elif rebalance[-1] == 'y':
    rebalance = relativedelta(years=int(rebalance[:-1]))
```

```python
data = web.get_data_yahoo(symbols, start = st-lookback, end=en)['Adj Close']
data.head()
```

```python
data_ret = data.pct_change().dropna()+1
data_ret.head()
```

```python
def mv_opt(R,C):
    n = len(R)
    def pf_var(W):
        return np.matmul(np.matmul(W,C),W)
    W = np.ones(n)/n
    e = np.ones(n)
    bnds = ((0.0,1.0),)*n
    #mu = 1+(0.07/252)
    cons = [#{'type' : 'ineq' , 'fun' : lambda w: np.dot(w,R) - mu}, # returns - mu >= 0
            {'type' : 'eq' , 'fun' : lambda w: np.sum(w) - 1}]
        #{'type':'ineq', 'fun': lambda x: np.sum(x, 1) - 1},   # col sum >= 1
        #{'type':'ineq', 'fun': lambda x: 1 - np.sum(x, 1)})  # col sum <= 1)
    solution = minimize(pf_var,W,method = 'SLSQP',bounds=bnds,constraints = cons)
    return solution.x.round(6)*100
R = np.array(np.mean(data_ret,axis=0))
C = data_ret.cov().values
mv_opt(R,C)
```

```python
def backtest(symbols,st,en,lb, rb):
    data = web.get_data_yahoo(symbols, start = st-lookback, end=en)['Adj Close']
    df = pd.DataFrame(columns = symbols)
    date = st-lookback
    while date < en:
        df.loc[data[date:].index[0]] = data[date:].values[0]
        date += rb
    data_ret = df.pct_change().dropna()+1
    
    col = [i+'_amt' for i in symbols]
    col.append('total_amt')
    col.append('portfolio_ret')
    col+= [i+'_weight' for i in symbols]
    result = pd.DataFrame(columns = col)
    T_amt = 100
    for index, row in df.iterrows():
        if st <= index:
            R = np.array(np.mean(data_ret[index-lb:index],axis=0))
            C = data_ret[index-lb:index].cov().values
            weight = mv_opt(R,C)
            amt = data_ret[index:index]*weight*T_amt/100
            for i in range(len(symbols)):
                s = symbols[i]
                result.loc[index,s+'_amt'] = amt[s].values[0]
                result.loc[index,s+'_weight'] = weight[i]
            T_amt_ = T_amt
            T_amt = amt.sum(axis = 1).values[0]
            result.loc[index,'total_amt'] = T_amt
            result.loc[index,'portfolio_ret'] = T_amt/T_amt_ - 1
    return result
backtest(symbols,st,en,lookback,rebalance)
```
